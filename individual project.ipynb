### research question2: will people feel much lack companionship with longer time spending on social media?

analysis:
    
1.variables: independent variable (x-axis):Social media time per day (continuous variable in hours). Dependent Variable(y-axis): Loneliness scale (categorical with three levels: hardly ever, some of the time, often).

2.Hypothesis test:
Null Hypothesis (Hâ): There is no association between the time spent on social media (in hours per day) and the loneliness scale rating.
Alternative Hypothesis (Hâ): There is an association between the time spent on social media and the loneliness scale rating, where higher social media time correlates with more frequent loneliness.
as there is a lot of data (with NA) so i want to use bootstrapping to improve the sample that help to find the p value.

3.reasons to use Hypothesis Testing:
Hypothesis testing provides a structured framework for making inferences about a population based on sample data. It allows researchers to decide whether to accept or reject the null hypothesis based on statistical evidence rather than subjective judgment.

```python
import pandas as pd
import numpy as np
from scipy.stats import norm

# Prepare the data
data = pd.DataFrame({
    'LONELY_ucla_loneliness_scale_companionship': [
        np.nan, np.nan, np.nan, np.nan, 'Some of the time', np.nan, np.nan, np.nan, 'Often', 
        np.nan, 'Some of the time', np.nan, np.nan, 'Hardly Ever', np.nan, np.nan, np.nan, np.nan, 
        'Hardly Ever', 'Some of the time', np.nan, 'Some of the time', np.nan, np.nan, 'Some of the time', 
        np.nan, np.nan, np.nan, np.nan, np.nan, 'Some of the time', 'Often', np.nan, np.nan, np.nan, 
        np.nan, 'Some of the time', np.nan, 'Some of the time', 'Hardly Ever', 'Often', np.nan, 
        np.nan, 'Some of the time'
    ],
    'CONNECTION_social_media_time_per_day': [
        '31-60 minutes per day', '2-3 hours per day', '2-3 hours per day', '1-2 hours per day', 
        '1-2 hours per day', '10-30 minutes per day', 'Less than 10 minutes per day', '1-2 hours per day', 
        '2-3 hours per day', '2-3 hours per day', 'More than 3 hours per day', '2-3 hours per day', 
        np.nan, '2-3 hours per day', '1-2 hours per day', '10-30 minutes per day', np.nan, 
        '10-30 minutes per day', np.nan, '1-2 hours per day', np.nan, np.nan, np.nan, '2-3 hours per day', 
        '1-2 hours per day', 'More than 3 hours per day', 'More than 3 hours per day', '31-60 minutes per day', 
        '10-30 minutes per day', np.nan, '10-30 minutes per day', 'More than 3 hours per day', 
        '10-30 minutes per day', np.nan, '10-30 minutes per day', '31-60 minutes per day', '1-2 hours per day', 
        '31-60 minutes per day', np.nan, '2-3 hours per day', '31-60 minutes per day', '10-30 minutes per day', 
        np.nan, '31-60 minutes per day'
    ]
})

# Map social media time to numeric values (for simplicity in hours)
social_media_map = {
    'Less than 10 minutes per day': 0.1,
    '10-30 minutes per day': 0.3,
    '31-60 minutes per day': 0.75,
    '1-2 hours per day': 1.5,
    '2-3 hours per day': 2.5,
    'More than 3 hours per day': 3.5
}
data['social_media_time_numeric'] = data['CONNECTION_social_media_time_per_day'].map(social_media_map)

# Drop rows with NA values
data = data.dropna(subset=['LONELY_ucla_loneliness_scale_companionship', 'social_media_time_numeric'])

# Step 1: Define bootstrap function
def bootstrap_mean_diff(data, group_col, value_col, group1, group2, n_iterations=1000):
    observed_data = data[data[group_col].isin([group1, group2])]
    group1_data = observed_data[observed_data[group_col] == group1][value_col]
    group2_data = observed_data[observed_data[group_col] == group2][value_col]
    observed_diff = group1_data.mean() - group2_data.mean()
    
    diffs = []
    combined_data = observed_data[value_col].values
    
    for _ in range(n_iterations):
        resample = np.random.choice(combined_data, size=len(combined_data), replace=True)
        resample_group1 = resample[:len(group1_data)]
        resample_group2 = resample[len(group1_data):]
        diff = np.mean(resample_group1) - np.mean(resample_group2)
        diffs.append(diff)
    
    diffs = np.array(diffs)
    p_value = np.sum(np.abs(diffs) >= np.abs(observed_diff)) / n_iterations
    return observed_diff, diffs, p_value

# Step 2: Perform bootstrap between "Hardly Ever" and "Often" groups
observed_diff, diffs, p_value = bootstrap_mean_diff(
    data, 'LONELY_ucla_loneliness_scale_companionship', 'social_media_time_numeric', 
    'Hardly Ever', 'Often', n_iterations=1000
)

print(f"Observed Difference in Means: {observed_diff}")
print(f"P-Value: {p_value}")
```

analysis result: Since the p-value of 0.812 is indeed much greater than 0.01, this provides strong evidence against rejecting the null hypothesis. Therefore, it can be concluded that, at least in the selected sample, there does not seem to be a clear correlation between social media use and feeling lonely. However, this does not mean that a correlation will not be found in other cases or larger samples.
So from the existing analysis, the time spent using social media does not have much impact on social happiness.

### research question3: will received a covid vaccine influence people's feeling about their life as a whole right now.

analysis:
    
1.variables: 
Independent Variable:COVID_vaccinated.
This is a categorical variable indicating the number of COVID-19 vaccine doses each person received.For linear regression, weâll encode these categories numerically, so that we can analyze how each dose level might impact life satisfaction.
Dependent Variable: WELLNESS_life_satisfaction.This is a continuous variable representing life satisfaction, likely on a scale from 1 to 10.As our outcome variable, it helps measure the potential impact of vaccination dose on overall life satisfaction.

2.Simple Linear Regression:
we use intercept: the predicted score of life satisfaction for unvaccinated individuals. Coefficient (slope): This indicates how much life satisfaction changes with each additional dose of the drug. 
also we should solve the NA.

3.Reasons to use Simple Linear Regression
Assumed Linear Relationship: We hypothesize that there may be a linear relationship between the number of vaccine doses and life satisfaction. For instance, people with more doses might feel more secure about their health, potentially improving their outlook on life.
Ease of Interpretation: Linear regression makes it easy to understand how an increase in dose count might affect life satisfaction, with the coefficient directly showing this relationship.
Statistical Significance: With linear regression, we can calculate a p-value to check if the number of vaccine doses has a significant effect on life satisfaction.

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt

# Load data into a DataFrame
data = pd.DataFrame({
    'COVID_vaccinated': [
        'Yes, three or more doses', 'Yes, three or more doses', 'Yes, three or more doses', 
        'Yes, three or more doses', 'Yes, one dose', 'Yes, three or more doses', 
        'Yes, two doses', 'Yes, three or more doses', 'Yes, one dose', 'Yes, three or more doses', 
        'No', 'No', 'NA', 'Yes, two doses', 'Yes, four doses', 'Yes, three or more doses', 
        'NA', 'Yes, four doses', 'No', 'Yes, two doses', 'NA', 'Yes, three or more doses', 
        'NA', 'Yes, two doses', 'No', 'Yes, three or more doses', 'Yes, four doses', 
        'Yes, two doses', 'NA', 'NA', 'Yes, two doses', 'No', 'Yes, five or more doses', 
        'NA', 'Yes, three or more doses', 'Yes, four doses', 'Yes, one dose', 'Yes, three or more doses', 
        'Yes, one dose'
    ],
    'WELLNESS_life_satisfaction': [
        2, 7, 5, 4, 8, 7, 5, 5, 1, 6, 7, 6, 8, 8, 6, 9, 7, 6, 5, 4, 'NA', 2, 'NA', 7, 1, 1, 
        8, 7, 9, 'NA', 1, 5, 8, 2, 6, 6, 6, 3, 2
    ]
})

# Remove rows where either column is 'NA'
data = data[(data['COVID_vaccinated'] != 'NA') & (data['WELLNESS_life_satisfaction'] != 'NA')]

# Encode vaccination status numerically
dose_mapping = {
    'No': 0,
    'Yes, one dose': 1,
    'Yes, two doses': 2,
    'Yes, three or more doses': 3,
    'Yes, four doses': 4,
    'Yes, five or more doses': 5
}
data['COVID_vaccinated'] = data['COVID_vaccinated'].map(dose_mapping)

# Convert life satisfaction to numeric values
data['WELLNESS_life_satisfaction'] = data['WELLNESS_life_satisfaction'].astype(float)

# Define X and Y variables
X = sm.add_constant(data['COVID_vaccinated'])  # Adding constant term for intercept
Y = data['WELLNESS_life_satisfaction']

# Fit the Simple Linear Regression model
model = sm.OLS(Y, X).fit()

# Plot data points and regression line
plt.figure(figsize=(8, 6))
plt.scatter(data['COVID_vaccinated'], data['WELLNESS_life_satisfaction'], color='blue', label='Data points')
plt.plot(data['COVID_vaccinated'], model.predict(X), color='red', label='Regression line')

# Adding labels and title
plt.xlabel('COVID-19 Vaccination Doses')
plt.ylabel('Life Satisfaction')
plt.title('Simple Linear Regression: COVID-19 Vaccination Doses vs Life Satisfaction')
plt.legend()
plt.show()
```
