The null hypothesis assumes that humans have no head tilt preference when kissing, meaning there is a 50% chance for either direction.Null Hypothesis:p=0.5 Alternative Hypothesis:p!=0.5,In the dataset:80 out of 124 couples tilted to the right.The observed proportion is:80/124≈0.645,Imagine a scenario where the null hypothesis is true, and each couple has a 50/50 chance of tilting left or right. This can be simulated using a large number of trials, where for each trial, we simulate 124 coin flips (representing the couples). In each simulation, calculate the proportion of "right tilts" (heads).The p-value is the probability of obtaining a result as extreme as, or more extreme than, the observed proportion (0.645) under the null hypothesis.Run many (e.g., 10,000) simulations of 124 coin flips.Count how often the simulated proportion of "right tilts" is greater than or equal to 0.645.
The p-value is the fraction of simulations that result in proportions as extreme or more extreme than the observed 0.645.


```python
import numpy as np

# Parameters for the simulation
n_couples = 124  # Number of couples
n_right_tilts = 80  # Observed number of right tilts
p_null = 0.5  # Null hypothesis probability (50/50 chance)
n_simulations = 10000  # Number of simulations

# Run the simulation: for each simulation, simulate n_couples coin flips (50% chance each)
simulations = np.random.binomial(n=n_couples, p=p_null, size=n_simulations)

# Compute the proportion of right tilts in each simulation
simulated_proportions = simulations / n_couples

# Calculate the observed proportion
observed_proportion = n_right_tilts / n_couples

# Compute the p-value: proportion of simulations where the result is as extreme or more extreme than the observed proportion
p_value = np.mean(simulated_proportions >= observed_proportion)
p_value
```

Null Hypothesis (H₀): The students are simply guessing, and the probability of correctly identifying whether milk or tea was poured first is equal to random guessing (50%).
Formally: H0:p=0.5, where p is the population proportion of correct guesses.
Informal interpretation: The students have no real ability to distinguish the pouring order, and any success is due to random chance.
Alternative Hypothesis (H₁): The students are not guessing randomly, and the proportion of correct identifications differs from 50%.
Formally: H1:p>0.5 (if we assume students have better-than-chance abilities) Informal interpretation: Students have some ability to identify the pouring order that is better than random guessing.
Quantitative Analysis
We will perform a formal hypothesis test to determine whether the observed result (49 out of 80 students correctly identifying the pouring order) provides sufficient evidence to reject the null hypothesis. Specifically, we will conduct a binomial test or a z-test for proportions, comparing the observed proportion of correct guesses to the 50% success rate expected under random guessing.

```python
from statsmodels.stats.proportion import proportions_ztest

# Observed data
correct_guesses = 49  # Number of correct guesses
total_students = 80  # Total number of students

# Calculate observed proportion
observed_proportion = correct_guesses / total_students
print(f"Observed proportion: {observed_proportion:.4f}")

# Set the null hypothesis proportion
null_hypothesis_proportion = 0.5

# Perform the z-test
z_stat, p_value = proportions_ztest(count=correct_guesses, nobs=total_students, value=null_hypothesis_proportion)

# Print the results
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the results
if p_value < 0.05:
    print("P-value is less than the significance level of 0.05, so we reject the null hypothesis. The students' ability is better than random guessing.")
else:
    print("P-value is greater than the significance level of 0.05, so we fail to reject the null hypothesis. The students' success may be due to random guessing.")
```
