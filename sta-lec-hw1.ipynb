Count: The number of non-null (non-missing) values for each variable. This represents the sample size available for the analysis of that particular variable.

Mean: The arithmetic average of the values for each variable. It is calculated by summing all the non-null values and dividing by the count (number of non-null values).

Standard Deviation (std): A measure of the dispersion or spread of the values around the mean. It indicates how much the individual values typically differ from the mean value. A higher standard deviation means more variability.

Minimum (min): The smallest value in the dataset for each variable.

25th Percentile (25%): Also known as the first quartile (Q1), it represents the value below which 25% of the data falls. In other words, 25% of the data points are less than or equal to this value.

50th Percentile (50%): Also known as the median, this is the middle value of the data. Half of the data points are less than or equal to this value, and half are greater than it.

75th Percentile (75%): Also known as the third quartile (Q3), it represents the value below which 75% of the data falls. In other words, 75% of the data points are less than or equal to this value.

Maximum (max): The largest value in the dataset for each variable.
